```python
# We have to convert the categorical data into the numerical data so we are using encoding techinques.
# Using label encoding to convert the "Gender" column into the numerical value.

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["Gender"]=le.fit_transform(data["Gender"])
data
```

```python
# Using label encoding to convert the column "state" into the numerical value.

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["State"]=le.fit_transform(data["State"])
data
```

```python
# Using label encoding to convert the column "City" into the numerical value.

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["City"]=le.fit_transform(data["City"])
data
```

```python
# Taking the required columns that are necessary as input for building the model and converting those values to array form


changed_input=data.iloc[:,0:4]
X=changed_input.values
```

```python
# Taking the output Column

output=data.iloc[:,4:]
```

```python
Y=output.values
```

### Step-4 :: Splitting the dataset into the train and test dataset.

```python
# Splitting the dataset into train set and test set.

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.5)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

### Step-4 :: Fitting the model-1(K-Nearest Neighbours)

```python
# Importing the requried model and evaluation metrics and here we are using "K-Nearest Neighbours" algorithm.

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Finding the best value for k to build the model

acc=[]
for i in range(1,21):
    neigh=KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train,Y_train)
    y_pred=neigh.predict(X_test)
    score=accuracy_score(Y_test,y_pred)
    acc.append(score)
```

```python
acc
```

```python
print("The best accuracy of the model is {0} with k-value {1}".format(max(acc),acc.index(max(acc))))
```

```python
# Creating the object for KNN with K-Value=3

neighbour=KNeighborsClassifier(n_neighbors=3)
```

```python
# Fitting the KNN model with k-value=3

neighbour.fit(X_test,Y_test)
```

```python
# Predict the y-values

y_pred=neighbour.predict(X_test)
```

```python
y_pred
```

### Step-5 :: Caluculating the accuracy score of the model.

```python
# Caluculating the accuracy Score

print("Accuracy Score:")
Accuracy_score=accuracy_score(Y_test,neighbour.predict(X_test))
print(Accuracy_score*100)
```
